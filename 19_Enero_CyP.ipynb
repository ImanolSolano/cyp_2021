#Slicing (Continuación) 

[Inicio : stop : incremento]

Notas...
1. El stop no es incluido. 
1. El incremento se puede modificar. 
1. Srporta indexado alternativo (negativos). 



```
frase="¿Cómo hacer a la gente más positiva?, muy facil, has que pierdan un electrón"
print(frase)
#Imprimir "ace" de la palabra "hacer"
print(frase[7:10:1])

#Valores por defecto
"""
Al no poner nada dentro del parentesis, obtiene los valores de [inicio de la
oración : Termino de la oración : salto de uno en uno]
"""
print(frase[::])
print(len(frase))
#Incremento
print("\n")
print(frase[::2]) #Contemplando toda la frase
print(frase[-1::-1]) #Inverso
print(frase[1::-1]) #Es una combinación de indexos.
print("\n")

nombre="Juan José"
print(nombre[1::-1])

primer=nombre[:4:]
print(primer) 

print(nombre[-9::1]) #Imprimir la frase completa con indexado negativo
```

#Métodos de la clase String (str)
En python, todo es un objeto (P.O.O.), eso significa que cada tipo de datos, trae consigo métodos (funciones) incluidas. 

Para identificar lo metodos disponibles, se puede unar una de dos funciones:
1. dir(str) #Te hace un resumen de los metodos de esa clase. 
2. help(str) #Información completa de los métodos de esa clase.  

```
nombre="Juan José"
print(nombre)
print(dir (str))
```

Los nombres de métodos que inician con doble guión bajo, significa que son métodos privados (P.O.O), y no tendrías porque usarlos como programador consumidor. 

Los demás métodos son públicos y son de libre us. 


```
print(nombre)
print(nombre.islower()) #Si es menor

nombre="PEDRO FERNANDEZ"
print(nombre.capitalize()) #Cambiar las letras a tipo oración es decir la primera con mayuscula y las demas en minuscula
print(nombre.endswith("DEZ")) #Comprobar si la terminación es la que se desea.
print(nombre.replace("E", "*")) #Remplazar algun espacio por algún caracter o letra especifica
ind_e=nombre.find("E") #Localizar una letra en especifico de toda la frase o cadena
print(ind_e)
nombre=nombre[:ind_e:]+ "*" + nombre[ind_e+1::] #Remplazar una letra en especifico
print(nombre)
print(nombre.replace("E", "@", 0)) #Remplazar una letra en especifico

nombre="PEDRO FERNANDEZ" 
print(nombre.lower()) #Para cambiar todas las letras a minusculas

nombre="josé"
print(nombre.upper()) #Para cambiar las letras a mayusculas

nombre= "josé josé, el principe de la canción"
print(nombre.count("josé")) #Para contar cuantas veces se repite algo en especifico surante la frase
```

```
#Métodos para procesar Strings

numeros="  2  ,    1,3,4,    5,  45         "
print(numeros)

#¿Cómo extraigo ls números de la cadena? 
#Paso 1, Eliminar los espacios al principio y al final con strip()
numeros=numeros.strip() #Eliminar los espacios del principio y del final
print(numeros)

#Paso 2 separo por comas
numeros=numeros.split(",") #Los separa por algun caracter en especial
print(numeros)

#Paso 3, con un for e indices limpiar cadenas de la lista
for indice in range (len(numeros)):
  numeros[indice]=numeros[indice].strip()
  print(numeros)

#Paso 4, convertirlos a tipo entero
for indice in range (len(numeros)):
  numeros[indice]=int(numeros[indice])
print(numeros)

```
